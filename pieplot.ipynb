```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv("Visadataset.csv")
df
```

```python
plt.pie(df["continent"].value_counts().values,labels=df["continent"].value_counts().keys())
plt.show()
```

#### autopct parameter : 

- autopct = "%1.1f%%"  
- autopct = "%matissa(main_value).precision(no of digits after the decimal point)"

- It is used to show the percentage of each unique categories on pieplot outoff 100%.

```python
plt.pie(df["continent"].value_counts().values,labels=df["continent"].value_counts().keys(),autopct="%1.1f%%")
plt.show()
```

```python
#### explode parameter : 

- It is used to seperate the categories in pieplot.
```

```python
plt.pie(df["continent"].value_counts().values,labels=df["continent"].value_counts().keys(),autopct="%1.1f%%",explode=[0.1 for i in range(6)])
plt.show()
```
