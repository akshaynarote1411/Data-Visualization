```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tips.csv")
df
```

```python
plt.boxplot(df["tip"])
plt.show()
```

```python
plt.boxplot(df["total_bill"])
plt.ylabel("Total Bill")
plt.title("Boxplot on Total Bill")
plt.show()
```

```python
sns.boxplot(df["tip"],color="green")
plt.show()
```

```python
sns.boxplot(x=df["total_bill"])
plt.title("Boxplot on Total Bill")
plt.show()
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind="box")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=df,kind="box")
plt.show()
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind="box",hue="time")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=df,kind="box",hue="sex")
plt.show()
```
