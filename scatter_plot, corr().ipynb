```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv("Visadataset.csv")
df
```

#### scatter() :

- It is applicable on continous variable to show the points on the graph.

```python
x= range(0,11)
y= range(0,11)
plt.scatter(x,y)
plt.plot(x,y)
```

```python
x = range(0,5)
sq = list(np.square(x))
plt.scatter(x,sq)
plt.plot(x,sq)
```

####  correlation(corr()) :

- corr() is the function which is used to show the relation between two columns.

- for corr(), we should only take continue varibales(numerical columns).

```python
df.corr()   # It will gives error because in df non-numericals columns also there, so corr() is not applicable on categorical columns.
```

```python
df.corr(numeric_only=True)   # numeric_only = True means it will take only numerical columns and apply corr().
```
