### Bar Plot :
- Used to represent the frequency of the categories from the variable on the graph.

#### Bar plot is applicable on two descrete varibles to represent the frequency. It required atleast two arguments one is on x-axis and other plot on y-axis.

- But we can also use Bar plot on one descrete varible to represent the frequency of that variable by plotting that variable unique categories on a axis and the count of that each category on other axis.

- we can identify unique categories of a variable as:
-  Syntax:   df["var_name"].value_counts().index --> It will gives the array of the indexes of the output(dataframe) of the value_counts of that variable
           - or
- df["var_name"].unique()  --> It will gives unique values.

- We can identify value counts of unique categories of a variable as:
-    Syntax :  df["var_name"].value_counts


```python
df["sex"].value_counts()
```

```python
df["sex"].value_counts().index
```

```python
df["sex"].value_counts().index.tolist()
```

```python
df["sex"].unique()
```

```python
plt.bar(df["sex"].unique(),df["sex"].value_counts())
plt.show()
```

```python
plt.bar(df["sex"],df["tip"])
```

```python
plt.bar(df["sex"])
```

### Stacked Bar  :
- It used when we have to show two bars on one bar only.

```python
plt.bar(df["sex"],df["total_bill"],label="Total_bill")
plt.bar(df["sex"],df["tip"],label="Tips")
plt.legend()
plt.show()
```

### UnStacked Bar :
- It is used when we want to show the two bars of same categories side by side.

```python
plt.bar(df["sex"],df["total_bill"],width=0.4,label="Total_bill")
plt.bar(df["sex"],df["tip"],width=0.4,label="Tip")
plt.xticks([0,1],df["sex"].unique().tolist())
plt.show()
```
