```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Visadataset.csv")
df
```

```python
print(df)  # It is DataFrame

print(df["continent"])  # It is a Series

print(df["continent"].values)   # It is Array means Raw Data.
```

```python
df["continent"].value_counts().values
```

```python
keys = df["continent"].value_counts().keys()
values = df["continent"].value_counts().values
```

```python
plt.bar(keys,values)
```

```python
shape=(10,4)
plt.figure(figsize=shape)
plt.bar(keys,values,color="red")
plt.xlabel("Class Names")
plt.ylabel("Class Frequency")
plt.title("Continent_Barchart")
plt.savefig("continent_bar.jpg")
plt.show()
```

```python
employee_education = df["education_of_employee"].value_counts().keys()
count_of_employee = df["education_of_employee"].value_counts().values
```

```python
plt.bar(employee_education,count_of_employee)
```

```python
plt.figure(figsize=(10,6))
plt.bar(employee_education,count_of_employee,color=["red","orange","green","blue"])
plt.xlabel("Employee Qualification")
plt.ylabel("Frequency of Emplyees")
plt.title("Employee Education BarPlot")
plt.savefig("employee_education.jpg")
plt.show()
```

```python
i="education_of_employee"

employee_education = df[i].value_counts().keys()
count_of_employee = df[i].value_counts().values


plt.figure(figsize=(10,6))
plt.bar(employee_education,count_of_employee,color=["red","orange","green","blue"])
plt.xlabel("Employee Qualification")
plt.ylabel("Frequency of Emplyees")
plt.title(f"{i} barchart")
plt.savefig(f"{i}.jpg")    # It is used to save the plot_image
plt.show()
```
